```python
import matplotlib.pyplot as plt

names = ["Arailym", "Mansur", "Ramazan", "Daniya", "Pankgea"]
scores = [85, 78, 92, 88, 75]

plt.plot(names, scores)
plt.xlabel("Students")
plt.ylabel("Scores")
plt.title("Student Scores")
plt.show()

plt.bar(names, scores)
plt.xlabel("Students")
plt.ylabel("Scores")
plt.title("Student Scores Bar Chart")
plt.show()
```
